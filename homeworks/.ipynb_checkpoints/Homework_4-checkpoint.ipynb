```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import f1_score
import missingno as msno
```

```python
## Import test and training data.
train = pd.read_csv('Costa_Rica_Data/train.csv')
test = pd.read_csv('Costa_Rica_Data/test.csv')
cols = train.columns
```

```python
print(train.shape)
print(test.shape)
```

## Data Cleaning

```python
## First I grabbed the test and training data and put them together in the dataset for cleaning.
ntrain = train.shape[0]
ntest = test.shape[0]
target = train['Target'].values
all_data = pd.concat((train, test)).reset_index(drop=True)
all_data.drop(['Target'],axis=1, inplace=True)
```

```python
## There are quite a few null values in certain columns, but less than 3% of the columns have null values between
## the train and test.
test.isnull().sum()[test.isnull().sum() > 0]
```

```python
## Find number of missing data in entire dataset.
train.isnull().sum()[train.isnull().sum() > 0]
```

```python
## I then used msno matrix to find the number of nulls in each column.
msno.matrix(train[['v2a1', 'v18q1', 'rez_esc', 'meaneduc', 'SQBmeaned']], color = (0.211, 0.215, 0.274))
plt.show()
```

```python
## We analyze the specific missing data for v2a1: this column is based on monthly rent and many of the values 
## are missing and it has high variance. Thus we look at its correspondence to monthly rent.
sns.countplot(all_data.loc[(pd.isnull(all_data['v2a1'])), 'tipovivi1'])
plt.title("Ownership")
plt.show()
```

```python
## Many of the people with these missing values already own their home, thus we can fill in those null values
## with 0.
all_data.loc[(pd.isnull(all_data['v2a1']) & all_data['tipovivi1'] == 1), 'v2a1'] = 0
all_data.loc[pd.isnull(all_data['v2a1']), 'v2a1'] = all_data['v2a1'].mean()
```

```python
## Values in v18q1 do not have 0s, thus we can assume these null values are 0.
all_data['v18q1'].unique()
```

```python
all_data.loc[(pd.isnull(all_data['v18q1']))] = 0
```

```python
## Fill in years behind school columns with 0, as it is difficult to change these column values.
all_data.loc[pd.isnull(all_data['rez_esc']), 'rez_esc'] = 0
```

```python
## Fill in average years of school with years of school.
all_data.loc[pd.isnull(all_data['meaneduc']), 'meaneduc'] = all_data.loc[pd.isnull(all_data['meaneduc']), 'escolari']
```

```python
## Fill in squared mean values with new mean values acquired.
all_data.loc[pd.isnull(all_data['SQBmeaned']), 'SQBmeaned'] = all_data.loc[pd.isnull(all_data['SQBmeaned']), 'meaneduc']**2
```

```python
## Work with dependency column with different data values.
all_data.groupby('dependency').size()
```

```python
all_data.isnull().sum()[all_data.isnull().sum() > 0]
```

```python
## We can replace dependency by the square root of the dependency squared, and store them as floats.
all_data.loc[(all_data['dependency']=='no'), 'dependency'] =  np.sqrt(all_data['SQBdependency'])
all_data.loc[all_data['dependency']=='yes', 'dependency'] = np.sqrt(all_data['SQBdependency'])
all_data['dependency'] = all_data['dependency'].astype('float16')

## We can do the same for edjefe and edjefa columns, replacing the 'no' with 0 years of education and 'yes' 
## with 4 years of education.
all_data.loc[all_data['edjefe']=='no', 'edjefe'] = 0
all_data.loc[all_data['edjefe']=='yes', 'edjefe'] = 4
all_data['edjefe'] = all_data['edjefe'].astype('uint8')
    
all_data.loc[all_data['edjefa']=='no', 'edjefa'] = 0
all_data.loc[all_data['edjefa']=='yes', 'edjefa'] = 4
all_data['edjefa'] = all_data['edjefa'].astype('uint8')
```

## Feature Engineering

```python
## Depending on certain values of the data, and because of the massive number of data columns, there are large 
## number of feature engineering columns we can grab from them.

## For this section, I grabbed several different techniques and feature engineering columns from different
## kernels online. 

def feature_engineer(x):
    ## *** Grab several different values based on age, and ability. *** ## 
    x['escolari_age'] = x['escolari'] / x['age'] ## years of schooling over age
    
    ## Convert values into integers
    x['refrig'] = x['refrig'].astype(int)
    x['computer'] = x['computer'].astype(int)
    x['television'] = x['television'].astype(int)
    x['mobilephone'] = x['mobilephone'].astype(int)
    x['v14a'] = x['v14a'].astype(int)
    x['v18q'] = x['v18q'].astype(int)
    x['epared1'] = x['epared1'].astype(int)
    x['epared2'] = x['epared2'].astype(int)
    x['epared3'] = x['epared3'].astype(int)
    x['etecho1'] = x['etecho1'].astype(int)
    x['etecho2'] = x['etecho2'].astype(int)
    x['etecho3'] = x['etecho3'].astype(int)
    
    x['eviv1'] = x['eviv1'].astype(int)
    x['eviv2'] = x['eviv2'].astype(int)
    x['eviv3'] = x['eviv3'].astype(int)
    x['abastaguadentro'] = x['abastaguadentro'].astype(int)
    x['abastaguafuera'] = x['abastaguafuera'].astype(int)
    x['abastaguano'] = x['abastaguano'].astype(int)
    x['abastaguano'] = x['abastaguano'].astype(int)
    x[['instlevel1', 'instlevel2', 'instlevel3', 'instlevel4', 'instlevel5', 'instlevel6', 'instlevel7', 'instlevel8', 'instlevel9']] = x[['instlevel1', 'instlevel2', 'instlevel3', 'instlevel4', 'instlevel5', 'instlevel6', 'instlevel7', 'instlevel8', 'instlevel9']].apply(pd.to_numeric)
    
    x['appliances'] = (x['refrig'] + x['computer'] + x['television'])
    
    ## Rent by specific household members background.
    x['rent_by_hhsize'] = x['v2a1'] / x['hhsize'] # rent by household size
    x['rent_by_people'] = x['v2a1'] / x['r4t3'] # rent by people in household
    x['rent_by_rooms'] = x['v2a1'] / x['rooms'] # rent by number of rooms
    x['rent_by_living'] = x['v2a1'] / x['tamviv'] # rent by number of persons living in the household
    x['rent_by_minor'] = x['v2a1'] / x['hogar_nin']
    x['rent_by_adult'] = x['v2a1'] / x['hogar_adul']
    x['rent_by_dep'] = x['v2a1'] / x['dependency']
    x['rent_by_head_educ'] = x['v2a1'] / (x['edjefe'] + x['edjefa'])
    x['rent_by_educ'] = x['v2a1'] / x['meaneduc']
    x['rent_by_numPhone'] = x['v2a1'] / x['qmobilephone']
    x['rent_by_gadgets'] = x['v2a1'] / (x['computer'] + x['mobilephone'] + x['v18q'])
    x['rent_by_num_gadgets'] = x['v2a1'] / (x['v18q1'] + x['qmobilephone'])
    x['rent_by_appliances'] = x['v2a1'] / x['appliances']
    
    x['under12'] = x['r4t1']/x['r4t3']
    x['under12_male'] = x['r4h1']/x['r4t3']
    x['under12_female'] = x['r4m1']/x['r4t3']
    x['Proportion_male'] = x['r4h3']/x['r4t3']
    x['Proportion_female'] = x['r4m3']/x['r4t3']
    
    x['tablet_density'] = x['v18q1'] / x['r4t3']
    x['phone_density'] = x['qmobilephone'] / x['r4t3']
    
    x['wall_qual'] = x['epared3'] - x['epared1']
    x['roof_qual'] = x['etecho3'] - x['etecho1']
    x['floor_qual'] = x['eviv3'] - x['eviv1']
    x['water_qual'] = x['abastaguadentro'] - x['abastaguano']
    
    x['house_qual'] = x['wall_qual'] + x['roof_qual'] + x['floor_qual']
    
    x['person_per_room'] = x['hhsize'] / x['rooms']
    x['person_per_appliances'] = x['hhsize'] / x['appliances']
    
    x['educ_qual'] = (1 * x['instlevel1']) + (2 * x['instlevel2']) + (3 * x['instlevel3']) + (4 * x['instlevel4']) + (5 * x['instlevel5']) + (6 * x['instlevel6']) + ( 7 * x['instlevel7']) + (8 * x['instlevel8']) + (9 * x['instlevel9'])
    x['educ_by_individual'] = x['escolari']/x['r4t3']
    x['educ_by_adult'] = x['escolari']/(x['r4t3'] - x['r4t1'])
    x['educ_by_child'] = x['escolari']/x['r4t1']
    
    x['max_educ'] = np.max(x[['edjefa','edjefe']])
    
    def reverse_label_encoding(row, df):
        for c in df.columns:
            if row[c] == 1:
                return int(c[-1])
            
    def rate_sanitary(row, df):
        c = df.columns.tolist()[0]
        
        if row[c] == 'sanitario2':
            return 3
        elif row[c] == 'sanitario3':
            return 2
        elif row[c] == 'sanitario5':
            return 1
        else:
            return 0
        
    def rate_cooking(row, df):
        c = df.columns.tolist()[0]
        
        if row[c] == 'energcocinar2':
            return 3
        elif row[c] == 'energcocinar3':
            return 2
        elif row[c] == 'energcocinar4':
            return 1
        else:
            return 0
        
    def rate_rubbish(row, df):
        c = df.columns.tolist()[0]
        
        if row[c] == 'elimbasu1':
            return 1
        elif row[c] == 'elimbasu2':
            return 2
        else:
            return 0
            
    x['sanitary'] = x.apply(lambda q: reverse_label_encoding(q, x[['sanitario1', 'sanitario2', 'sanitario3', 'sanitario5', 'sanitario6']]), axis=1)
    x['cooking'] =  x.apply(lambda q: reverse_label_encoding(q, x[['energcocinar1', 'energcocinar2', 'energcocinar3', 'energcocinar4']]), axis=1)
    x['rubbish'] = x.apply(lambda q: reverse_label_encoding(q, x[['elimbasu1', 'elimbasu2', 'elimbasu3', 'elimbasu4', 'elimbasu5', 'elimbasu6']]), axis=1)
    x['region'] = x.apply(lambda q: reverse_label_encoding(q, x[['lugar1', 'lugar2', 'lugar3', 'lugar4', 'lugar5', 'lugar6']]), axis=1)
    
    x['sanitary_i'] = x.apply(lambda q: rate_sanitary(q, x[['sanitary']]), axis = 1)
    x['cooking_i'] = x.apply(lambda q: rate_cooking(q, x[['cooking']]), axis = 1)
    x['rubbish_i'] = x.apply(lambda q: rate_rubbish(q, x[['rubbish']]), axis = 1)
    
    x['zone'] = x['area1'] - x['area2']

    x.replace([np.inf, -np.inf], 0, inplace = True)
    x.fillna(0, inplace = True)

feature_engineer(all_data)
```

```python
## I then got the dummy variables for the model, which grew the data from 142 columns to over 10000 columns!! Maybe
## it is a little too many, but overall I don't think it will have a major effect on my values.
all_data.isnull().sum()[all_data.isnull().sum() > 0]
print(all_data.columns)
all_data1 = pd.get_dummies(all_data)
print(all_data1.columns)
```

## Modeling using Gridsearch

```python
## Randomforest model using Gridsearch.
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report, roc_auc_score

x = all_data1.loc[:(ntrain-1),cols].drop(['Id','idhogar'], axis = 1)
x_test = all_data1.loc[ntrain:,cols].drop(['Id','idhogar'], axis=1)
target = train['Target']
```

```python
## Use GridSearch to find best parameters for RandomForestClassifier
param_dictionary = {"n_estimators": [100,200,300,400], 'max_depth':[3,4,5]}
clf = RandomForestClassifier(n_jobs=-1, class_weight='balanced')
gs = GridSearchCV(clf, param_dictionary, n_jobs=1, verbose=2, cv=2)
gs.fit(x,target)
```

```python
## The Classifier is different each time, but in this case it chose max_depth of 3 and n_estimator of 300. Thus
## I will predict the data on this.
clf = gs.best_estimator_
clf.fit(x,target)
```

```python
## We predict the train and test data, and get about 35% accuracy which is pretty bad!! I must have some type of
## error in the data, but still, I would think this model would predict more accurately for me...
train_predictions = clf.predict(x)
test_preds = clf.predict(x_test)
print('% Correct:',(target == train_predictions).sum()/len(target) * 100)
```

```python
## Here is the classification report for my specific data: these values are EXTREMELY different between the two
## different classifications. My recall is extremely low, which shows this test is not very reliable overall
## I believe the data skewed has a large impact on the test/training data.
cr = classification_report(target,train_predictions)
print(cr)
```
